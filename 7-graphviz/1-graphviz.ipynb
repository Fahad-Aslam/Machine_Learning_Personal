#### Creating a Simple Graph
Now, let's create a simple example of a graph using Graphviz in Python. We'll start with a basic directed graph with nodes and edges

````python
from graphviz import Digraph

# Create a new directed graph
graph = Digraph()

# Add nodes
graph.node('A')
graph.node('B')
graph.node('C')

# Add edges
graph.edge('A', 'B')
graph.edge('B', 'C')
graph.edge('C', 'A')

# Save the graph to a file and render it
# graph.render('simple_graph', format='png', cleanup=True)

graph
````

##### Explanation:
- ```from graphviz import Digraph```: This line imports the ```Digraph``` class from the ```graphviz``` package, which allows us to create directed graphs.
- ```graph = Digraph():``` This line initializes a new directed graph.
- ```graph.node('A')```, ```graph.node('B')```, ```graph.node('C')```: These lines add nodes labeled 'A', 'B', and 'C' to the graph.
- ```graph.edge('A', 'B')```, ```graph.edge('B', 'C')```, ```graph.edge('C', 'A')```: These lines add directed edges between nodes.
- ```graph.render('simple_graph', format='png', cleanup=True)```: This line saves the graph to a file named 'simple_graph.png' in PNG format and cleans up temporary files after rendering.

---

#### Viewing the Graph
After running this script, you should find a file named simple_graph.png in your current directory. Open this file to see the rendered graph.

#### Customizing Graphs
You can customize graphs by changing node colors, shapes, edge styles, and more. Here's an example:

````python
from graphviz import Graph

# Create a new undirected graph
graph = Graph()

# Add nodes with custom attributes
graph.node('A', color='red', shape='box')
graph.node('B', color='blue', shape='ellipse')
graph.node('C', color='green', shape='circle')

# Add edges with custom attributes
graph.edge('A', 'B', color='purple', penwidth='2.0')
graph.edge('B', 'C', color='orange', style='dashed')

# Save and render the graph
# graph.render('custom_graph', format='png', cleanup=True)

# Display the graph in the IPython console or Jupyter Notebook
graph
````

# graphviz with sklearn

##### Step 1: Install Required Libraries
- First, make sure you have both scikit-learn and graphviz installed. You can install them using pip if you haven't already:
```bash
pip install scikit-learn graphviz
```

---
#### Step 2: Import Libraries
- Now, import the necessary libraries in your Python script:

````python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier, export_graphviz
import graphviz
````

---
#### Step 3: Load Data
- For this example, let's use the Iris dataset, which is included in scikit-learn. We'll load the dataset and create a decision tree classifier.

````python
# Load the Iris dataset
iris = load_iris()
X = iris.data
y = iris.target

# Create a decision tree classifier
clf = DecisionTreeClassifier()
clf.fit(X, y)
````

#### Step 4: Visualize the Decision Tree
- Now, we'll use export_graphviz to generate a Graphviz representation of the decision tree and then visualize it.

````python
# Generate DOT data
dot_data = export_graphviz(clf, out_file=None, 
                           feature_names=iris.feature_names,  
                           class_names=iris.target_names,  
                           filled=True, rounded=True,  
                           special_characters=True)  

# Create and display the graph
graph = graphviz.Source(dot_data)  
graph.render("iris_decision_tree", format='png')  # Save tree to file
graph  # Display tree in Jupyter Notebook or IPython
````

#### Explanation:
- ```export_graphviz```: This function generates a Graphviz representation of the decision tree model (clf) in DOT format.
- ```clf```: The trained decision tree classifier.
- ```out_file=None:``` Specifies where to save the output DOT file. Setting it to None returns the DOT data as a string.
- ```feature_names:``` Names of the features (e.g., sepal length, sepal width, petal length, petal width).
- ```class_names:``` Names of the target classes (e.g., setosa, versicolor, virginica).
- ```filled```, ```rounded```, ```special_characters```: These parameters control the appearance of the tree nodes.
- ```graphviz.Source```: This class represents the DOT source code and allows us to render and display the graph.
- ```graph.render("iris_decision_tree", format='png'):``` This line saves the decision tree visualization to a PNG file named iris_decision_tree.png.

---
#### Viewing the Decision Tree
After running the script, you should find a file named iris_decision_tree.png in your current directory, containing the visualization of the decision tree. Open this file to view the tree diagram.

Further Customization
You can further customize the appearance of the decision tree by adjusting the parameters passed to export_graphviz, such as filled, rounded, class_names, feature_names, etc. Experiment with these parameters to create visualizations that best suit your needs.
