##### Working with images in tf.keras involves two main parts:  
-> loading and preparing the images <br> -> and then using those images to train a model.  

We'll cover both parts with some fun examples.

---

### Imagine we're training a model to recognize cats and dogs!

#### 1. Loading Images:
There's a special function in tf.keras.preprocessing.image called load_img that can grab an image from your computer.  Here's how to use it:

```python
from tensorflow.keras.preprocessing.image import load_img

# Replace 'path/to/your/image.jpg' with the real path to your image
cat_img = load_img('greenML.png') 
cat_img
```

This code loads the image and stores it in a variable called cat_img. But it's not quite ready for our model yet.  Computers work with numbers, and images are stored as a bunch of tiny colored squares called pixels.  We need to convert the image into a format our model understands.

---

#### 2. Preparing Images:

Another function from tf.keras.preprocessing.image called img_to_array can help us with this.  It takes the image loaded by load_img and converts it into an array of numbers.  We can also resize the image to a specific size our model expects. Here's how to do both:

```python
from tensorflow.keras.preprocessing.image import load_img, img_to_array

# This resizes the image to be 150 pixels tall and 150 pixels wide
cat_img = img_to_array(load_img('greenML.png', target_size=(4, 4)))
cat_img
```

Now cat_img is a numpy array ready for our model.  Remember to resize all your images to the same size!

---

#### 3. Putting it all together:

We can use these steps to load and prepare a bunch of cat and dog images for our model.  Here's an example loop that goes through two folders, one with cats and one with dogs:

```python
from tensorflow.keras.preprocessing.image import load_img, img_to_array
import os
cat_data = []
dog_data = []

# Replace 'path/to/cats/' and 'path/to/dogs/' with the real paths to your folders
for filename in os.listdir('greenML.png'):
  img = img_to_array(load_img('greenML.png' + filename, target_size=(150, 150)))
  cat_data.append(img)

# for filename in os.listdir('path/to/dogs/'):
#   img = img_to_array(load_img('path/to/dogs/' + filename, target_size=(150, 150)))
#   dog_data.append(img)
```

This loop loads all the cat images into a list called cat_data and dog images into dog_data.  Now you have a collection of images ready to be used to train your cat vs dog model!

---

#### Bonus! Showcasing the Image:

While training, you might want to see what your images look like after being converted into arrays. You can use a library called matplotlib to display the images. Here's how to show the first cat image we loaded:

```python
import matplotlib.pyplot as plt

plt.imshow(cat_data[0])
plt.show()
```

This will pop up a window showing the first image from our cat_data list.
